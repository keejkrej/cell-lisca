```python
import h5py
import matplotlib.pyplot as plt
```

```python
h5_path = "/Users/jack/Documents/20251216_cell_000.h5"

def print_info(name, obj):
    data = obj[:] if isinstance(obj, h5py.Dataset) else None
    print(name)
    if data is not None:
        print(data.shape)

with h5py.File(h5_path, "r") as f:
    f.visititems(print_info)
```

```python
with h5py.File(h5_path, "r") as f:
    print(f["fov_0/cell_0/sequence_0/channels"][:])
```

```python
def plot_h5(h5_path, frames):
    with h5py.File(h5_path, "r") as f:
        fig, ax = plt.subplots(5, len(frames), figsize=(4 * len(frames), 10))
        for a in ax.flat:
            a.axis("off")
        for i, frame in enumerate(frames):
            ax[0,i].imshow(f["fov_0/cell_0/sequence_0/data"][frame,0])
            ax[1,i].imshow(f["fov_0/cell_0/sequence_0/data"][frame,1])
            ax[2,i].imshow(f["fov_0/cell_0/sequence_0/data"][frame,2])
            ax[3,i].imshow(f["fov_0/cell_0/sequence_0/cell_masks"][frame])
            ax[4,i].imshow(f["fov_0/cell_0/sequence_0/nuclei_masks"][frame])
        plt.show()
```

```python
plot_h5(h5_path, [0, 10, 20, 30, 40, 50, 60])
```
